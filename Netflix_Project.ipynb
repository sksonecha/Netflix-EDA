```python
act=pd.DataFrame()
act=mydata['cast'].str.split(',',expand=True).stack()
act=act.to_frame()
act.columns=['Actors']
act
```

```python

topactors.plot(kind='barh',y='No of performances',x='Actors',color='green')
plt.title("Actors with maximum number of Performances")
```

Conclusion - Anupam Kher is the most featured actor on Netflix for the time gap of 2002 - 2021.
